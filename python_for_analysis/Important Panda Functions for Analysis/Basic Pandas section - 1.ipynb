### Position-based Indexing and Slicing with iloc[]

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete")
```

```python
summer # Year are considered 0 column index, for indexing [row, column]
```

```python
summer.info()
```

```python
summer.Year.unique()
```

```python
summer.City.unique()
```

```python
summer.Sport.unique()
```

```python
summer.Discipline.unique()
```

#### Selecting Rows with iloc[]

```python
summer.iloc[0] # one column and one row return
```

```python
type(summer.iloc[0]) 
```

```python
summer.iloc[1]
```

```python
summer.iloc[-1]
```

```python
summer.iloc[[1, 2, 3]]
```

```python
summer.iloc[1:4]
```

```python
summer.iloc[:5] # first five rows
```

```python
summer.iloc[-5:] # last five rows
```

```python
summer.iloc[:] # all rows
```

```python
summer.iloc[[2, 45, 5467]]
```

#### Indexing/Slicing Rows and Columns with iloc[]

```python
summer.head(10)
```

```python
summer.iloc[0, 4] # on zero rows and four of column i.e: Hungery
```

```python
summer.iloc[[0,4]]
```

```python
summer.iloc[0, :3]
```

```python
summer.iloc[0, [0, 2, 5, 7]]  # select the specific value of column
```

```python
summer.iloc[34:39, [0, 2, 5, 7]]
```

#### Selecting Columns with iloc[]

```python
summer.iloc[:, 4]
```

```python
summer.iloc[:, 4].equals(summer.Country)
```

```python
summer["Country"]
```
